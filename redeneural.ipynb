vamos começar analisando o modelo de programção classico. Para isso, vamos criar uma função que receberá um número (x) e retornará o nnumero recebido multiplicado por 2 e subtraído de 1.

y = 2x - 1

```python
def multiplicaPorDoisSubtraiUm(x):
  y = (x*2) - 1
  return y
```

```python
multiplicaPorDoisSubtraiUm(8)
```

Acima utilizamos o metodo tradicional de programção, onde desenvolvemos o script e passamos os dados para a maquina processar e nos mostrar o resultado. Agora, vamos passar apenas as entradas e os resultados e a máquina irá gerar a regra.

```python
import tensorflow as tf
model = tf.keras.Sequential([tf.keras.layers.Dense(units=1, input_shape=[1])])
#unit -> unidade de "neuronio" / input_shape -> unidade de entradas
```

Uma vez que o neuronio foi criado, é necessário passarmos os parametros de configuração

```python
model.compile(optimizer="sgd", loss="mean_squared_error") #optimizer de pesos para minizar a função de perda / loss, função de perda de custo
```

Hora da mágica, vamos passar os valores de entrada e os resultados para o neuronio ser treinado.


```python
import numpy as np
X = np.array([-1.0, 0.0, 3.0, 4.0, 5.0]) #flutuante tem precisão maior
y = np.array([-3.0, -1.0, 5.0, 7.0,9.0])
```

```python
model.fit(X, y, epochs=500)
```

```python
model.predict([8])
```
